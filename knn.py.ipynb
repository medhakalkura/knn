```python
import numpy as np
import pandas as pd
import math
import operator

data=pd.read_csv("nbadataset.csv")
data=data[['Age','G','GS','MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','eFG%','FT','FTA','FT%','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PS/G','Player','Tm','Pos']]
print(data)

def distance(data1,data2,length):
    distan=0
    for x in range(length):
        distan+=np.square(data1[x]-data2[x])
    return(np.sqrt(distan))

def knn(trainingset,testdata,k):
    dist={}
    sorteddist={}
    neigh=[]
    length=testdata.shape[1]
    for x in range(len(trainingset)):
        d=distance(trainingset.iloc[x],testdata,length)
        dist[x]=d[0]
    distsort=sorted(dist.items(),key=operator.itemgetter(1))
    
    classname={}
    for x in range(k):
        neigh.append(distsort[x][0])
    for x in range(len(neigh)):
        i = trainingset.iloc[neigh[x]][-1]
        if i in classname:
            classname[i]=classname[i]+1
        else:
            classname[i]=1
    sortedclass=sorted(classname.items(),key=operator.itemgetter(1),reverse=True)
    return(sortedclass[0][0],neigh)

testset = [[20,32,2,10,3.4,10.8,0.31,1,2.5,0.405,3.4,8.2,0.55,0.458,3.4,4.0,0.752,0.5,3,3.5,2,0.6,0.3,1.6,2.3,13.9]]
test = pd.DataFrame(testset)
k=1
clas,neigh=knn(data,test,k)
print(clas)
print(neigh) 

k=3
clas,neigh=knn(data,test,k)
print(clas)
print(neigh)

k=5
clas,neigh=knn(data,test,k)
print(clas)
print(neigh)
```
