```python
import numpy as np
import pandas as pd
import math
import operator
import random
data=pd.read_csv("nbadataset.csv")
data=data[['Age','G','GS','MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','eFG%','FT','FTA','FT%','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PS/G','Player','Tm','Pos']]
trainingset= data.sample(frac=0.67,random_state=200)
testset=data.drop(trainingset.index)
print(trainingset)
print(len(trainingset))
print(testset)
print(len(testset))

def distance(data1,data2,length):
    distan=0
    for x in range(length):
        distan+=np.square(data1[x]-data2[x])
    return(np.sqrt(distan))

def knn(trainingset,testdataset,k):
    
    predicted=[]
    for y in range(len(testdataset)):
        dist={}
        sorteddist={}
        neigh=[]
        length=26
        for x in range(len(trainingset)):
            d=distance(trainingset.iloc[x],testdataset.iloc[y],length)
            dist[x]=d
        distsort=sorted(dist.items(),key=operator.itemgetter(1))
    
        classname={}
        for x in range(k):
            neigh.append(distsort[x][0])
        for x in range(len(neigh)):
            i = trainingset.iloc[neigh[x]][-1]
            if i in classname:
                classname[i]=classname[i]+1
            else:
                classname[i]=1
        sortedclass=sorted(classname.items(),key=operator.itemgetter(1),reverse=True)
        predicted.append(sortedclass[0][0])
        print(sortedclass[0][0])
        print(neigh)
    
    
    
    correct=0
    for x in range(len(testdataset)):
        if testdataset.iloc[x][-1] is predicted[x]:
            correct+=1
    return (correct/float(len(testdataset)))*100.0

k=1
acc=knn(trainingset,testset,k) 
print("accuracy is ")
print(acc)

k=3
acc=knn(trainingset,testset,k) 
print("accuracy is ")
print(acc)

```
